```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Transformed_Housing_Data2.csv')
data.head()
```

```python
#SCALING THE DATASET
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Y = data['Sale_Price'] #putting target var as y
X = scaler.fit_transform(data.drop(columns = ['Sale_Price']))
X = pd.DataFrame(data = X, columns = data.drop(columns = ['Sale_Price']).columns)
X.head()
```

```python
#CHECKING AND REMOVING MULTICOLLINEARITY
X.corr()
```

```python
## pair of independent variables with correlation greater than 0.5
k = X.corr()
z = [[str(i),str(j)] for i in k.columns for j in k.columns if (k.loc[i,j] >abs(0.5))&(i!=j)]
z, len(z)
```

```python
# Importing Variance_inflation_Factor funtion from the Statsmodels
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data = X

## Calculating VIF for every column
VIF = pd.Series([variance_inflation_factor(vif_data.values, i) for i in range(vif_data.shape[1])], index = vif_data.columns)
VIF
```

removing all those variables with VIF more than 5

if one variable is removed, there is a good chance that the VIF of other variables will also go down.

Process:
1. Calculate VIF
2. Any variable higher than 5?
3. If yes, Remove variable with the highest VIF and recalculate
4. If no, stop

```python
# which has the max vif rn
VIF[VIF == VIF.max()]
```

```python
#.index[0] returns only the first index (i.e. the name here in this case)
VIF[VIF == VIF.max()].index[0]
```

```python
#Now to remove the variable
def MC_remover(data):
  #first we calculate the vif of all the things
  vif = pd.Series([variance_inflation_factor(data.values, i) for i in range(data.shape[1])], index = data.columns)
  #then we check if the max vif is greater than 5
  if vif.max() > 5:
    print(vif[vif == vif.max()].index[0],'has been removed')
    #we remove the variable
    data = data.drop(columns = [vif[vif == vif.max()].index[0]])
    return data
  else:
    print('No Multicollinearity present anymore')
    return data
```

```python
for i in range(7): #we are putting range as 7 cuz we found out before that there were 7 variables with vif>5
   vif_data = MC_remover(vif_data)

vif_data.head()
```

```python
# Calculating VIF for remaining columns
VIF = pd.Series([variance_inflation_factor(vif_data.values, i) for i in range(vif_data.shape[1])], index = vif_data.columns)
VIF, len(vif_data.columns)
```

# Training the model

```python
#TRAIN/TEST SET
X = vif_data #scaled data with removed multicollinearity 
Y = data['Sale_Price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 101)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
#we could create our own gradient descent code for this but there is another easier method.
from sklearn.linear_model import LinearRegression
lr = LinearRegression(normalize = True) #this will set the values of all independant variables such that lr will not have to find the value of intercept c.
lr.fit(x_train, y_train) #lr.fit() is the one wihich does the gradient descent
```

```python
lr.coef_
#we are finding the coefficient of the indepndant variable rn.
#these coeff are nothing but m1, m2, m3 ... etc from y = m0 + m1x1 + m2x2... etc
#since we have normalised the data, m0 = 0
```

```python
#now generating predictions over the test set
predictions = lr.predict(x_test)
```

```python
#we are now going to evaluate the entire test dataset (x test and y test) by calculating the R2
lr.score(x_test, y_test)
```

Now we are done with step 8, we are almost done.

moving on to step 9 where we verify the assumptions of linear regression

```python
#plotting the residuals
residuals = predictions - y_test
#new dataframe for residuals
residual_table = pd.DataFrame({'residuals':residuals,
                    'predictions':predictions})
#arranging the residual data
residual_table = residual_table.sort_values( by = 'predictions')
```

why do we need to arrange the residual values according to the predictions which are arranged in ascending order?


yk we didnt use any independat variable while plotting the residual plot of the mean regression model (wo upar wala)

because no independant variable was used while building the mean regression model

but now we are looking at multiple independant variables

how do we visualise the residual plot where we have multiple variables?
- the predicted values (or the y_hat values) represent the effect/outcome of all the independant variables combined.

```python
z = [i for i in range(int(residual_table['predictions'].max()))]
k = [0 for i in range(int(residual_table['predictions'].max()))]
```

```python
plt.figure(dpi = 130, figsize = (17,7))

plt.scatter( residual_table['predictions'], residual_table['residuals'], color = 'red', s = 2)
# s is the size of the dots
plt.plot(z, k, color = 'green', linewidth = 3, label = 'regression line')
plt.ylim(-800000, 800000)
plt.xlabel('fitted points (ordered by predictions)')
plt.ylabel('residuals')
plt.title('residual plot')
plt.legend()
plt.show()
```

```python
#frequency table of residual. like how many error terms per residual value.
plt.figure(dpi = 100, figsize = (10,7))
plt.hist(residual_table['residuals'], color = 'red', bins = 200)
plt.xlabel('residuals')
plt.ylabel('frequency')
plt.title('distribution of residuals')
plt.show()
```

# Model Coefficient

```python
#we create a dataframe for every column name and their respective coeff
coefficients_table = pd.DataFrame({'column': x_train.columns,
                                  'coefficients': lr.coef_})
coefficient_table = coefficients_table.sort_values(by = 'coefficients')
```

```python
# now we can find out which variable plays the most role and influences the sale price the most in a positive way.
plt.figure(figsize=(8, 6), dpi=120)
x = coefficient_table['column']
y = coefficient_table['coefficients']
plt.barh( x, y)
plt.xlabel( "Coefficients")
plt.ylabel('Variables')
plt.title('Normalized Coefficient plot')
plt.show()
```

this says that area of house increases the sale price
